# Data Cleaning

```python
product_details["color"]=product_details["color"].fillna(product_details["color"].ffill())
```

```python
product_details.isnull().sum()
```

```python
product_details["category"]=product_details["category"].replace({"???":"unknown"})
```

```python
product_details["supplier"]=product_details["supplier"].replace({"suppliera":"supplier_A","supplierb":"supplier_B","supplierc":"supplier_C","supplierd":"supplier_D"})
```

```python
product_details
```

```python
customer_details["gender"]=customer_details["gender"].replace({"???":"Other"})
```

```python
customer_details
```

```python
sales_details["customer_id"]=sales_details["customer_id"].fillna("unknown")
```

```python
sales_details["discount"]=sales_details["discount"].fillna(0)
```

```python
sales_details.isnull().sum()
```

```python
sales_details
```

# Data Merging


```python
data=sales_details.merge(product_details,on="product_id",how="left")
```

```python
data=data.merge(customer_details,on="customer_id",how="left")
```

```python
data=data.merge(store_details,on="store_id",how="left")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data["returned"]=data["returned"].replace({0:"No",1:"Yes"})
```

```python
col_name=["category","color","size","season","supplier","age","gender","city","store_name","region",]
for i in col_name:
    data[i]=data[i].fillna(data[i].mode()[0])
    print(i)
```

```python
col_nam=["cost_price","list_price","store_size_m2"]
for i in col_nam:
    data[i]=data[i].fillna(data[i].mean())
    print(i)
```

```python
data["date"]=pd.to_datetime(data["date"],format='%Y-%m-%d')
```

```python
data.isnull().sum()
```

```python
data
```

```python
data[data.duplicated()]
```

```python
data["sales_amount"]=data["list_price"]*data["quantity"]
```

```python
data["discount_amount"]=data["sales_amount"]*data["discount"]
```

```python
data["Net_amount"]=data["sales_amount"]-data["discount_amount"]
```

```python
data["profit"]=data["Net_amount"]-(data["cost_price"]*data["quantity"])
```

```python
data
```

# Visualization

```python
copy=data.copy()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8,4))
plt.hist(copy["age"])
```

```python
plt.figure(figsize=(5,4))
copy["year"]=copy["date"].dt.year.astype(int)
vis=copy.groupby("year")["Net_amount"].sum().sort_values(ascending=False).reset_index()
sns.lineplot(x="year",y="Net_amount",data=vis,color="gray")
plt.title("total revenue by year")
```

```python
plt.figure(figsize=(5,4))
vis=copy.groupby("year")["profit"].sum().reset_index()
sort=vis.sort_values(by="profit",ascending=False)
sns.lineplot(x="year",y="profit", data=sort )
plt.title("profit by year")
plt.show()
```

```python
copy["month_name"]=copy["date"].dt.month_name()
vis=copy.groupby("month_name")["Net_amount"].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,4))
sns.lineplot(x="month_name",y="Net_amount",data=vis,marker="o",color="green")
plt.title("total revenue by months")
```

```python
highest sale month has may and then june. low revenue generated month has february. insight mostly sale in hot days 
```

```python
vis=copy.groupby("category")["quantity"].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(4,4))
plt.barh(vis["category"],vis["quantity"],color="skyblue")
plt.gca().invert_yaxis()
plt.title("popular product by quantity")
plt.xlabel("Quantity")
plt.ylabel("Category")
```
